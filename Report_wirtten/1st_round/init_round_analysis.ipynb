```python
import pandas as pd
import numpy as np
```

# Initialization

```python
init = pd.read_csv('initialization.txt', sep = ',')
```

```python
y_init = pd.read_csv('init_1st_raw.csv', header = None)
```

```python
y_init[y_init > 1] = 1
```

```python
y_init
```

```python
y_init = y_init.values
```

```python
y_init = y_init.reshape(-1)
```

```python
y_init = np.round(y_init, 3)
```

```python
init['Cell viability'] = y_init
```

```python
init
```

```python
init.to_csv("initialization_1st_results.txt", header=True, index=None, sep=',', mode='w')
```
