```python
import matplotlib.pyplot as plt
```

```python
mean = [0.746434819, 0.774588668, 0.853867339, 0.877961205]
std = [0.015643456, 0.039997739, 0.021810935, 0.021336131]
label = ['(+)-Griseofulvin', '(S)-(+)-Camptothecin', 'Chloramphenicol', 'Metformin']
```

```python
plt.figure(figsize=(12, 8), dpi=100)
plt.errorbar(label, mean, std, linestyle='-', marker='^',color = 'blue')
plt.title('Cell Viability in Positive Control Wells', fontsize = 16)
plt.xticks(label, fontsize= 15)
plt.yticks(fontsize= 13)
plt.show()
```
