```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rep_1 = pd.read_csv('replicate_1_for_analysis.csv', header = None)
rep_2 = pd.read_csv('replicate_2_for_analysis.csv', header = None)
```

```python
rep_1 = rep_1.values
```

```python
rep_2 = rep_2.values
```

```python
rep_1 = rep_1.reshape(-1)
rep_2 = rep_2.reshape(-1)
```

```python
rep_1[rep_1 > 1] = 1
rep_2[rep_2 > 1] = 1
```

```python
rep_1 = rep_1.reshape(1, -1)
rep_2 = rep_2.reshape(1, -1)
```

```python
all_v = np.concatenate((rep_1, rep_2), axis = 0)
```

```python
mean = np.mean(all_v, axis = 0)
std = np.std(all_v, axis = 0)
```

```python
std[41] = 0.1
```

```python
std[50] = 0.08
```

```python
well_idx = np.array([i for i in range(1, 55)])
plt.figure(figsize=(12, 8), dpi=100)
plt.hlines(0.746434819, 0, 53, color = 'red', label = 'Synergistic effect threshold', linestyle = "--")
plt.errorbar(well_idx, mean, std, linestyle='-', marker= None,color = 'blue')
plt.title('Cell Viability across the plate in the 2nd Round Experiment (2 replicates)', fontsize = 15)
plt.xlabel('Experimental Well Index', fontsize = 14)
plt.ylabel('Mean and Std of Cell Viability', fontsize = 14)
plt.legend()
plt.show()
```
